```python
a = open('class8am.csv','w')
a.write("SN,Name,Age,Address,Phone No\n")
n = int(input("Enter no of row = "))
for i in range(1,n+1):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    add = input("Enter add = ")
    phone=int(input("Enter add = "))
    info = f"{i},{name},{age},{add},{phone}\n"
    a.write(info)
a.close()
```
